# Before doing clustering we have to find what are the optimal K(cluster) points :using elbow method:

```python
#imporing Keans......
from sklearn.cluster import KMeans
sse = []
k_rng = range(1,10)
for k in k_rng:
    km=KMeans(n_clusters=k)
    km.fit(dataframe_2[['Province','Faculty of Graduation','Donated?','Past Traveler Y/N','Preferred Language','Gender']])
    sse.append(km.inertia_)
```

```python
sse
```

#Here, we can see that elbow shape is taking place at point.2......Means 2 clusters

```python
plt.xlabel('k')
plt.ylabel('sum od squarred error')
plt.plot(k_rng,sse)
```

```python
km = KMeans(n_clusters=2)
```

```python
yy_predict = km.fit_predict(dataframe_2[['Province','Faculty of Graduation','Donated?','Past Traveler Y/N','Preferred Language','Gender']])
```

```python
dataframe_2['cluster'] = yy_predict
dataframe_2
```

```python
km.cluster_centers_
```

```python
dt1 = dataframe_2[dataframe_2.cluster==0]
dt2 = dataframe_2[dataframe_2.cluster==1]
```
